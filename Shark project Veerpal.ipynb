```python

# Clean column names (lowercase, replace spaces, regex)
df.columns = df.columns.map(lambda c: re.sub(r'[^A-Za-z0-9_]+', '', c.strip().lower().replace(" ", "_")))
```

```python
# Strip whitespace in string columns
df = df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
```

```python
# Remove duplicated rows
df = df.drop_duplicates()
```

```python

# Drop useless columns (Unnamed columns)
df = df.drop(columns=[c for c in df.columns if c.startswith("unnamed")], errors='ignore')
```

```python
# Convert "age" to numeric using regex to extract numbers
if "age" in df.columns:
    df["age"] = df["age"].astype(str).str.extract(r'(\d+)')
    df["age"] = pd.to_numeric(df["age"], errors='coerce')

# Fill numeric missing with mean & median examples
if "age" in df.columns:
    df["age"] = df["age"].fillna(df["age"].median())   # median example

if "year" in df.columns:
    df["year"] = pd.to_numeric(df["year"], errors='coerce')
    df["year"] = df["year"].fillna(df["year"].mean())  # mean example
```

```python
# -----------------------
# DATE CLEANING
# -----------------------
if "date" in df.columns:
    df["date"] = df["date"].apply(lambda x: pd.to_datetime(x, errors="coerce"))
```

```python
# -----------------------
# TEXT CLEANING WITH REGEX
# -----------------------
# Remove weird characters from name
if "name" in df.columns:
    df["name"] = df["name"].apply(lambda x: re.sub(r'[^A-Za-z\s]+', '', str(x)))

# Standardize sex field (M/F/Unknown)
if "sex" in df.columns:
    df["sex"] = df["sex"].apply(lambda s: re.sub(r'[^MFmf]', '', str(s)).upper())
    df["sex"] = df["sex"].replace("", "Unknown")
```
